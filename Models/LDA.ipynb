```python
# Importação das bibliotecas necessárias
import yfinance as yf
from pycaret.regression import *
```

```python


# Baixar dados das ações da B3
ticker = 'PETR4.SA'
data = yf.download(ticker, start="2020-01-01", end="2023-01-01")
data['Target'] = (data['Adj Close'].shift(-1) > data['Adj Close']).astype(int)
data.dropna(inplace=True)  # Removendo valores NaN que ocorrem devido ao shift

# Preparando a coluna de atraso para ser um recurso
data['Lagged_Price'] = data['Adj Close'].shift(1)
data.dropna(inplace=True)  # Remover quaisquer valores NaN

# Selecionando colunas necessárias
data = data[['Lagged_Price', 'Target']]
```

```python
from pycaret.classification import *

# Setup do PyCaret
exp_cls = setup(data, target='Target', session_id=123, train_size=0.8, data_split_shuffle=True)

# Comparando modelos para encontrar o melhor, assegurando-se de que a métrica de classificação esteja correta
best_model = compare_models(sort='Accuracy')  # Substitua 'Accuracy' pela métrica desejada confirmada anteriormente
```
